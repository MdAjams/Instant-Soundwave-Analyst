```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data load
df=pd.read_csv("mxmh_survey_cleaned.csv")
df.head()
```

# 1Ô∏è‚É£ Music Effect on Mental State (Overall)
#### Distribution of Satisfaction Responses

    üé∂ Let's see how many people believe that music truly helps their mental health!

```python
df.info()
```

```python
# Music effect on emotions
plt.figure(figsize=(7, 4))
sns.countplot(data=df, x='music_effects', palette='Set1')
plt.title("Does Music Help Your Mental State?")
plt.xlabel("Music Effect")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45)
plt.show()

# Overall Distribution of Music Effects
plt.figure(figsize=(7, 5))
sns.countplot(data=df, x='music_effects', palette='cool')
plt.title("Impact of Music on Mental State üé∂üß†")
plt.xlabel("Music Effects")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45)
plt.show()
```

#  Analyze Music Effects by Age Group
     Let's explore how different age groups perceive the emotional benefits of music!

```python
# Create Age Groups
bins = [12, 18, 25, 35, 50, 70]
labels = ['Teen', 'Young Adult', 'Adult', 'Middle Aged', 'Senior']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

# Plot
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='age_group', hue='music_effects', palette='Spectral')
plt.title("Impact of Music on Mental State by Age Group üéß")
plt.xlabel("Age Group")
plt.ylabel("Number of Respondents")
plt.legend(title="Music Effects")
plt.show()
```

# Analyze Music Effects by Primary Streaming Service
    üé∂ Do Spotify, YouTube, or Apple Music users feel differently about music's impact? Let's compare!



```python
# Top Streaming Services
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='primary_streaming_service', hue='music_effects', palette='Set2')
plt.title("Impact of Music Based on Streaming Platforms üì±üéµ")
plt.xlabel("Primary Streaming Service")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45)
plt.legend(title="Music Effects")
plt.show()
```

# Analyze Hours per Day vs Music Effects
    ‚è≥ Does listening to music for more hours relate to feeling more emotionally better?

```python
# Average Listening Hours Based on Music Effects
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='music_effects', y='hours_per_day', palette='Pastel1')
plt.title("Music Listening Time vs Mental Satisfaction ‚è≥üé∂")
plt.xlabel("Music Effects")
plt.ylabel("Hours Per Day")
plt.show()
```

# Correlation with Anxiety, Depression, Insomnia, OCD
    üîó How closely is the emotional effect of music connected to mental health issues like anxiety and depression?

```python
# Encode 'music_effects' into numeric values
df['music_effects_num'] = df['music_effects'].map({'Helps a lot': 3, 'Helps somewhat': 2, 'Neutral': 1, 'Does not help': 0})

# Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df[['music_effects_num', 'anxiety', 'depression', 'insomnia', 'ocd']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Music Satisfaction and Mental Health Indicators üîóüß†")
plt.show()
```
